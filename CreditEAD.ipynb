```python
def dataspecs(filename,target):
    name =[x for x in globals() if globals()[x] is filename][0]
    print("Current Dataframe Name is %s" % name)
    print("---------------------------------------------------------------")
    print(filename.info())
    print("  ")
    print("The dimensions of  %s" % name,"are",filename.shape)
    print("  ")
    print(filename.dtypes)
    print("  ")
    rows=filename.shape[0]
    print("  ")
    columns=filename.shape[1]
    print("  ")
    print(filename.head())
    print("  ")
    print(filename.tail())
    print("  ")
    desc=filename.describe(include='all')
    print("  ")
    print("Summary Statistics for %s" % name,"are",desc)
    target_list=filename[target].unique()
    print("  ")
    print(target_list)
    print("---------------------------------------------------------------")
    print("---------------------------------------------------------------")
    print("------------------------End of %s" % name,"----------------------------")
    print("                        ")
#def dataspecs(filename,target):
dataspecs(bpop,"CREDIT_TYPE")
dataspecs(bsamp10,"CREDIT_TYPE")
```
